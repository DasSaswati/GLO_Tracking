# JPL Flight Data
    Azimuth and elevation angles given in degrees
    Resolution of azimuth and elevation data appears to be ~0.6 degrees
    Looks like coarse pointing system kept balloon approx +/- 12 degrees in azimuth
    One of the most abrupt coarse pointing corrections has a slope of ~.65 deg/sec
    

```python
%matplotlib inline
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
cwd = os.getcwd()
```

```python
################################################################################
base_size=15
mpl.rcParams['legend.fontsize'] = base_size
mpl.rcParams['figure.figsize'] = (10,7.5)
mpl.rcParams['figure.titlesize']=base_size+5
mpl.rcParams['xtick.labelsize']=base_size
mpl.rcParams['ytick.labelsize']=base_size
mpl.rcParams['font.size']=base_size
mpl.rcParams['axes.titlesize']=base_size
mpl.rcParams['axes.labelsize']=base_size
mpl.rcParams['lines.markersize'] = 5           # markersize, in points
mpl.rcParams['legend.markerscale'] = 1     # line width in points
mpl.rcParams['lines.markeredgewidth'] = 0.2 # the line width around the marker symbol
mpl.rcParams['lines.linewidth'] = 1.5
#####################################
```

```python
file_loc = cwd+"/eng16271.out"
data = pd.read_csv(file_loc,header=2,delim_whitespace=True)
mean_t_diff = data['tsecs'].diff().mean()
data = data.set_index('tsecs')
data_mean_10 = data.rolling(center=True,window=10).mean()
data_mean_50 = data.rolling(center=True,window=50).mean()
data_mean_100 = data.rolling(center=True,window=100).mean()

mask_1 = (data.index > 62558) & (data.index < 90009)
mask_2 = (data.index > 62000) & (data.index < 70000)
mask_3 = (data.index > 63779) & (data.index < 63796)
min_azi_rate=63794
mask_4 = (data.index > min_azi_rate-200) & (data.index < min_azi_rate+200)
```

```python
plt.plot(data.loc[:,'AZIMUTH_Angle'],label='Azimuth')
plt.plot(data.loc[:,'ELEVATION_Angle'],label='Elevation')
plt.xlabel('Time Elapsed (seconds)')
plt.ylabel('Degrees')
plt.title('JPL Flight Data: Entire Flight')
plt.legend()
```

```python
plt.plot(data.loc[mask_1,'AZIMUTH_Angle'],label='Azimuth')
plt.xlabel('Time Elapsed (seconds)')
plt.ylabel('Degrees')
plt.title('JPL Flight Data')
plt.legend()
```

```python
plt.plot(data.loc[mask_2,'AZIMUTH_Angle'],label='Azimuth')
plt.plot(data.loc[mask_3,'AZIMUTH_Angle'],color='r',label='Steep Slope\napprox 0.65 deg/sec')
plt.xlabel('Time Elapsed (seconds)')
plt.ylabel('Degrees')
plt.title('JPL Flight Data')
plt.legend()
```

```python
plt.plot(data.loc[mask_3,'AZIMUTH_Angle'],color='r',label='Azimuth\nslope approx 0.65 deg/sec')
plt.xlabel('Time Elapsed (seconds)')
plt.ylabel('Degrees')
plt.title('JPL Flight Data: Zoomed in on a "Steep" slope')
plt.legend()
```

```python
#calculate stats for azimuth angular rates
std_10 = data_mean_10.loc[mask_1,'AZIMUTH_Angle'].diff().std()
std_50 = data_mean_50.loc[mask_1,'AZIMUTH_Angle'].diff().std()
std_100 = data_mean_100.loc[mask_1,'AZIMUTH_Angle'].diff().std()

#divide by average time between samples to get deg/sec angular rate
azi_rate_std_10 = std_10/mean_t_diff
azi_rate_std_50 = std_50/mean_t_diff
azi_rate_std_100 = std_100/mean_t_diff
```

```python
#1-sigma and 3-sigma values for azimuth angular rates
print('filter window=10:\n 1-sigma= ',round(azi_rate_std_10,4),'deg/sec',
      '\nfilter window=50:\n 1-sigma= ',round(azi_rate_std_50,4),'deg/sec',
      '\nfilter window=100:\n 1-sigma= ',round(azi_rate_std_100,4),'deg/sec')
print('')
print('filter window=10:\n 3-sigma= ',round(3*azi_rate_std_10,4),'deg/sec',
      '\nfilter window=50:\n 3-sigma= ',round(3*azi_rate_std_50,4),'deg/sec',
      '\nfilter window=100:\n 3-sigma= ',round(3*azi_rate_std_100,4),'deg/sec',)
```

```python
#Attempt to find the percentage of flight time that coarse pointing is engaged
#Assume that coarse pointing is engaged when the when the azimuth angular rate (deg/sec) is below a certain threshold
plt.plot(data_mean_50.loc[mask_1,'AZIMUTH_Angle'],label='Rolling mean - 50 samples')
diff_50_scaled = 20*data_mean_50.loc[mask_1,'AZIMUTH_Angle'].diff()/mean_t_diff
plt.plot(diff_50_scaled,'o',label='Diff of filter_50')
plt.plot(diff_50_scaled[diff_50_scaled < -1.2],'o',label='Diff of filter_50 below -1.2 deg')
plt.xlabel('Time Elapsed (seconds)')
plt.ylabel('Degrees')
plt.title('JPL Flight Data')
plt.legend()
```

```python
#calculate number of samples below threshold:
c_engaged = sum(n < -1.2 for n in (20*data_mean_50.loc[mask_1,'AZIMUTH_Angle'].diff()/mean_t_diff).values.flatten())

#Total number of samples
total = len(data_mean_50.loc[mask_1,:])

#Percent of time with coarse pointing engaged:
print(100*c_engaged/total)
```
